```python
#自定义一块画布，并命名为fig,画布大小为figsize:(18,6),分辨率dpi=1600
fig = plt.figure(figsize=(18,6),dpi=1600)#figsize : a tuple (width, height) in inches
alpha_bar_chart = 0.55

#将不同大小、不同种类的图像画在一块画布上

#分块1 将当前画布分成6块，ax1是第一行第一列
#绘制一个条形图，表示幸存or not
ax1 = plt.subplot2grid((2,3),(0,0))
df.Survived.value_counts().plot(kind='bar',alpha = alpha_bar_chart)#alpha表示图表的透明度，alpha越大，颜色越深
ax1.set_xlim(-1,2)#x轴的显示范围是-1~2，这样能使（survived=0）（survived=1）显示在图中间
plt.title("Ditribution of Survival(Survived=1)")

#分块2
ax2 = plt.subplot2grid((2,3),(0,1))
label0=[]
label1=[]
for i in df.index:
    if df.Survived[i]==1:
        label1.append(i)
    elif df.Survived[i]==0:
        label0.append(i)
plt.scatter(df.Age[label0],[0 for i in label0],marker='x',color='r',label='Not Survived')
plt.scatter(df.Age[label1],[1 for i in label1],marker='o',color='g',label='Survived')
plt.title("Age-Survived")
plt.legend(loc='best')
ax2.set_ylim(-1,2)
#plt.show()

#分块3
ax3 = plt.subplot2grid((2,3),(0,2))
c1=[]
c2=[]
c3=[]
for i in df.index:
    if df.Pclass[i]==1:
        c1.append(i)
    elif df.Pclass[i]==2:
        c2.append(i)
    elif df.Pclass[i]==3:
        c3.append(i)
plt.scatter(c1,[1 for i in c1],marker='o',color='r')
plt.scatter(c2,[2 for i in c2],marker='o',color='g')
plt.scatter(c3,[3 for i in c3],marker='o',color='b')
plt.title("index-Pclass")
#plt.show()

#分块4 Age Distribution within class
ax4 = plt.subplot2grid((2,3),(1,0),colspan=2)
#print(df.Age[df.Pclass==1][1])
#df.Age[df.Pclass==1]的类型是pandas.series df.Age[df.Pclass==1][7] 如果7号乘客乘坐1等客舱，那么这个式子表示他的年龄
df.Age[df.Pclass==1].plot(kind='kde',color='r',label='1st')#‘kde’ : Kernel Density Estimation plot 高斯图
df.Age[df.Pclass==2].plot(kind='kde',color='g',label='2nd')
df.Age[df.Pclass==3].plot(kind='kde',color='b',label='3rd')
plt.title("Age Distribution within class")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend(loc='best')

#分块5
ax5 = plt.subplot2grid((2,3),(1,2))
df.Embarked.value_counts().plot(kind='barh')
plt.ylabel("Embarked Position")
plt.title("#Passengers-Embarked")
plt.show()
```

formula:
----
    - '~'<=>'='
    - C()表示其内部变量是一个categorical variable(有固定范围的变量，比如性别限制在男女之间)

```python
from patsy import dmatrices

formula = 'Survived~C(Pclass)+Age+C(Sex)+SibSp+Parch+Fare+C(Embarked)'
#y,x = dmatrices(formula, data=df, return_type='dataframe')
Y_train,X_train = dmatrices(formula,data=df,return_type='matrix')
print(type(Y_train),type(X_train))
#print(Y_train)
X_train=np.array(X_train)
print(X_train[:30])
print(type(X_train))
```

```python
usedDf = df.drop(['PassengerId','Name'],axis=1)
usedDf
```

将dataFrame变成narray:
=================
    - https://jingyan.baidu.com/article/a17d52853f879e8099c8f240.html
    - df.iloc[[1,2,4],[0,2]]提取df的1，2，4行，0，2列
    - df.iloc[3]提取dataframe的第四行数据，类型为Series

```python
print(usedDf.columns)
X_data = np.array(usedDf.iloc[:,1:])
Y_data = np.array(usedDf.iloc[:,0])

print(X_data)

for item in X_data:
    if item[1] == 'male':
        item[1] = 1
    else:
        item[1] = 0
        
    if item[-1] == 'S':
        item[-1] = 0
    elif (item[-1] == 'Q'):
        item[-1] = 1
    if item[-1] == 'C':
        item[-1] = 2

print(X_data[1][-1])
print(X_data.shape)
Y_data.reshape((712,1))
print(Y_data.shape)
```

将数据集分为训练集和测试集
===================

```python
randPerIdx = np.random.permutation(X_data.shape[0])
print(randPerIdx[:20])
X_train = X_data[randPerIdx[:534],:]
X_test = X_data[randPerIdx[534:],:]
Y_train = Y_data[randPerIdx[:534]]
Y_test = Y_data[randPerIdx[534:]]
print(X_train.shape)
print(Y_train.shape)
```

```python
from keras.layers import Input,Dense
from keras.models import Model
```

```python
num_of_examples , num_of_feats = X_train.shape
print(num_of_feats,',',num_of_examples)
```

建立模型
========

```python
inputs = Input(shape=(7,))
x = Dense(30,activation='relu')(inputs)
x = Dense(14,activation='relu')(x)
x = Dense(7,activation='relu')(x)
x = Dense(7,activation='relu')(x)
predictions = Dense(1,activation='sigmoid')(x)

model = Model(inputs=inputs,outputs=predictions)
model.compile(optimizer='adagrad',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=1000,batch_size=32)
```

```python
score = model.evaluate(X_test, Y_test, batch_size=32)
print(score)#loss,accuracy
```
